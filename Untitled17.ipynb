```python
from transformers import pipeline
```

```python
a=pipeline("sentiment-analysis")
```

```python
gen=pipeline("text-generation")
```

```python
text=input("Enter the text: ")
t=print(gen(text,max_length=200))
```

```python
print(t[0].get('generated_text'))
```

```python
n=pipeline('ner')
```

```python
text=input("Enter the text: ")
result=n(text)
result
```

```python
qa=pipeline("question-answering")
```

```python
c='''Bity bought some butter,but the butter was bitter butter,
so,Bitty bought some better butter,to make bitter butter'''
```

```python
q="what does Bitty bought?"
```

```python
result=qa({"question":q,'context':c})
print(result)
```

```python
s=pipeline("summarization")
```

```python
t='''On a quiet hill surrounded by whispering pines,
a lonely lantern lit the porch of an old cottage.
Every evening, a little girl named Mira placed
the lantern there, hoping it would guide her
father home from sea. One stormy night, as
thunder cracked and rain blurred the world,
a shadowy figure appeared at the edge of the path.
The light from the lantern flickered in the wind,
and Mira’s heart raced — until a familiar voice
called her name, carrying through the storm like a promise kept.'''
```

```python
result=s(t)
result[0].get("summary_text")
```
