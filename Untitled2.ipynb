```python
import pandas as pd
import seaborn as sns
```

```python
dt=pd.read_csv('/content/large_student_pass_fail_dataset.csv')
```

```python
dt.describe()
```

```python
dt.info()
```

```python
dt['pass_fail'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
```

```python
dt['pass_fail']=l.fit_transform(dt['pass_fail'])
```

```python
dt['pass_fail'].unique()
```

```python
dt
```

```python
from sklearn.model_selection import train_test_split
```

```python
dt.corr()
```

```python
x=dt[['hours_studied','attendance_percentage','assignments_submitted']]
```

```python
y=dt['pass_fail']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
xtest
```

```python
sns.pairplot(dt,hue='pass_fail')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
l=LogisticRegression()
```

```python
l.fit(xtrain,ytrain)
```

```python
ypred=l.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```
